## Task 1: Divide and Conquer - Binary Search
Implementation and performance analysis.

```python
import time
import matplotlib.pyplot as plt
import numpy as np

def binary_search(arr, low, high, x):
    if high >= low:
        mid = (high + low) // 2
        if arr[mid] == x:
            return mid
        elif arr[mid] > x:
            return binary_search(arr, low, mid - 1, x)
        else:
            return binary_search(arr, mid + 1, high, x)
    else:
        return -1

# Performance Measurement
input_sizes = [1000, 5000, 10000, 50000, 100000, 500000]
times = []
for size in input_sizes:
    arr = list(range(size))
    start_time = time.time()
    binary_search(arr, 0, len(arr) - 1, size - 1)
    times.append(time.time() - start_time)

plt.figure(figsize=(10, 6))
plt.plot(input_sizes, times, marker='o', label='Binary Search Time')
plt.xlabel('Input Size')
plt.ylabel('Time (s)')
plt.title('Binary Search Complexity: O(log n)')
plt.grid()
plt.legend()
plt.show()
```

## Task 2: Sorting Performance Comparison
Comparing Merge Sort and Quick Sort.

```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        L = arr[:mid]
        R = arr[mid:]
        merge_sort(L)
        merge_sort(R)
        i = j = k = 0
        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1
        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1
        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1

def partition(arr, low, high):
    pivot = arr[high]
    i = low - 1
    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1

def quick_sort(arr, low, high):
    if low < high:
        pi = partition(arr, low, high)
        quick_sort(arr, low, pi - 1)
        quick_sort(arr, pi + 1, high)

import time
import matplotlib.pyplot as plt
import numpy as np

# Timing
input_sizes = [100, 500, 1000, 2000, 5000]
merge_times = []
quick_times = []

for size in input_sizes:
    arr = np.random.randint(0, 10000, size).tolist()
    
    start_time = time.time()
    merge_sort(arr.copy())
    merge_times.append(time.time() - start_time)
    
    start_time = time.time()
    quick_sort(arr.copy(), 0, len(arr) - 1)
    quick_times.append(time.time() - start_time)

plt.figure(figsize=(10, 6))
plt.plot(input_sizes, merge_times, label='Merge Sort', marker='o')
plt.plot(input_sizes, quick_times, label='Quick Sort', marker='x')
plt.xlabel('Input Size')
plt.ylabel('Time (s)')
plt.title('Sorting Performance: Merge vs Quick Sort')
plt.legend()
plt.grid()
plt.show()
```

## Task 3: Greedy Algorithms - Fractional Knapsack
Optimal resource allocation using a Greedy strategy.

```python
class Item:
    def __init__(self, value, weight):
        self.value = value
        self.weight = weight
        self.ratio = value / weight

def fractional_knapsack(capacity, items):
    items.sort(key=lambda x: x.ratio, reverse=True)
    total_value = 0.0
    for item in items:
        if capacity >= item.weight:
            capacity -= item.weight
            total_value += item.value
        else:
            total_value += item.value * (capacity / item.weight)
            break
    return total_value

# Sample Run
items = [Item(60, 10), Item(100, 20), Item(120, 30)]
capacity = 50
print(f"Max Value in Fractional Knapsack: {fractional_knapsack(capacity, items)}")
```

## Task 4: Dynamic Programming - 0/1 Knapsack
Solving the 0/1 Knapsack problem using Dynamic Programming.

```python
def knapsack_01(W, wt, val, n):
    dp = [[0 for x in range(W + 1)] for x in range(n + 1)]
    for i in range(n + 1):
        for w in range(W + 1):
            if i == 0 or w == 0:
                dp[i][w] = 0
            elif wt[i-1] <= w:
                dp[i][w] = max(val[i-1] + dp[i-1][w-wt[i-1]], dp[i-1][w])
            else:
                dp[i][w] = dp[i-1][w]
    return dp[n][W]

# Performance/Output
val = [60, 100, 120]
wt = [10, 20, 30]
W = 50
n = len(val)
print(f"Max Value in 0/1 Knapsack: {knapsack_01(W, wt, val, n)}")
```

## Task 5: Shortest Path - Dijkstra's Algorithm (Greedy Strategy)
Finding the shortest path in a graph.

```python
import heapq

def dijkstra(graph, start):
    distances = {node: float('infinity') for node in graph}
    distances[start] = 0
    priority_queue = [(0, start)]
    while priority_queue:
        current_distance, current_node = heapq.heappop(priority_queue)
        if current_distance > distances[current_node]:
            continue
        for neighbor, weight in graph[current_node].items():
            distance = current_distance + weight
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(priority_queue, (distance, neighbor))
    return distances

# Example Graph
graph = {
    'A': {'B': 1, 'C': 4},
    'B': {'A': 1, 'C': 2, 'D': 5},
    'C': {'A': 4, 'B': 2, 'D': 1},
    'D': {'B': 5, 'C': 1}
}
print(f"Shortest paths from A: {dijkstra(graph, 'A')}")
```

## Summary and Reflections

### Complexity Table
| Algorithm | Paradigm | Time Complexity (Best) | Time Complexity (Worst) |
|---|---|---|---|
| Binary Search | Divide & Conquer | O(1) | O(log n) |
| Merge Sort | Divide & Conquer | O(n log n) | O(n log n) |
| Quick Sort | Divide & Conquer | O(n log n) | O(n^2) |
| Fractional Knapsack | Greedy | O(n log n) | O(n log n) |
| 0/1 Knapsack | Dynamic Programming | O(nW) | O(nW) |
| Dijkstra's | Greedy | O(E log V) | O(E log V) |

### Observations
1. **Recursion Depth**: Algorithms like Quick Sort can hit the maximum recursion depth for large inputs if the pivot selection is not optimal.
2. **Practical Performance**: While Quick Sort has a worse worst-case complexity than Merge Sort, its average-case performance and smaller constant factors often make it faster in practice for randomized data.
